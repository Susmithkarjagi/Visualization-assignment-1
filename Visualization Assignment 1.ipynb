Q1: What is Matplotlib? Why is it used? Name five plots that can be plotted using the Pyplot module of
Matplotlib.

Matplotlib is a data visualization library for Python programming language. It provides a wide range of customizable static, animated, and interactive visualizations in Python.

Matplotlib is used to plot data in a variety of formats, such as line graphs, scatter plots, bar plots, histograms, and more. It is a powerful tool for exploring and presenting data in a clear and concise manner.

Here are five plots that can be plotted using the Pyplot module of Matplotlib:

1. Line plot: It is a plot of data points connected by straight lines to show the trend in the data.
2. Scatter plot: It is a plot of data points on a two-dimensional plane, where each point represents a value for two different variables.
3. Bar plot: It is a plot of rectangular bars with heights or lengths proportional to the values they represent. It can be used to compare values across different categories.
4. Histogram: It is a graphical representation of the distribution of a dataset. It is used to visualize the underlying frequency distribution of a continuous variable.
5. Pie chart: It is a circular statistical graphic, which is divided into slices to illustrate numerical proportion. It is used to show the proportion of different categories in a dataset.

Q2: What is a scatter plot? Use the following code to generate data for x and y. Using this generated data
plot a scatter plot.
import numpy as np
np.random.seed(3)
x = 3 + np.random.normal(0, 2, 50)
y = 3 + np.random.normal(0, 2, len(x))
Note: Also add title, xlabel, and ylabel to the plot.

A scatter plot is a graph that uses dots to represent data points on a two-dimensional plane, with one variable on the x-axis and the other variable on the y-axis. It is used to show the relationship between two variables and to identify patterns or trends in the data.

To generate a scatter plot using the provided code, we first need to import the necessary libraries and generate the data for x and y using the NumPy library.

```python
import numpy as np
import matplotlib.pyplot as plt

np.random.seed(3)
x = 3 + np.random.normal(0, 2, 50)
y = 3 + np.random.normal(0, 2, len(x))
```

Here, we have used the np.random.normal() function to generate two sets of random data with a mean of 3 and a standard deviation of 2. We have generated 50 data points for x and the same number of data points for y.

Now, we can plot the scatter plot using the plt.scatter() function from the Matplotlib library.

```python
plt.scatter(x, y)
plt.title('Scatter Plot')
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.show()
```

Here, we have used the plt.scatter() function to plot the scatter plot with x as the x-axis and y as the y-axis. We have also added a title, xlabel, and ylabel to the plot using the plt.title(), plt.xlabel(), and plt.ylabel() functions, respectively.

Finally, we have used the plt.show() function to display the scatter plot.

Q3: Why is the subplot() function used? Draw four line plots using the subplot() function.
Use the following data:
import numpy as np
For line 1: x = np.array([0, 1, 2, 3, 4, 5]) and y = np.array([0, 100, 200, 300, 400, 500])

For line 2: x = np.array([0, 1, 2, 3, 4, 5]) and y = np.array([50, 20, 40, 20, 60, 70])

For line 3: x = np.array([0, 1, 2, 3, 4, 5]) and y = np.array([10, 20, 30, 40, 50, 60])

For line 4: x = np.array([0, 1, 2, 3, 4, 5]) and y = np.array([200, 350, 250, 550, 450, 150])


The subplot() function in Matplotlib is used to create multiple plots within a single figure. It is useful when we want to display multiple plots side-by-side or in a grid layout. The subplot() function takes three arguments: the number of rows, the number of columns, and the index of the current subplot.

For example, if we want to create a figure with two rows and three columns of subplots, we can use the following code:

```python
import matplotlib.pyplot as plt

fig, axs = plt.subplots(nrows=2, ncols=3)
```

This will create a figure with six subplots arranged in two rows and three columns.

Now, using the provided data, we can create four line plots using the subplot() function. Here's the code:

```python
import numpy as np
import matplotlib.pyplot as plt

# Data for line 1
x1 = np.array([0, 1, 2, 3, 4, 5])
y1 = np.array([0, 100, 200, 300, 400, 500])

# Data for line 2
x2 = np.array([0, 1, 2, 3, 4, 5])
y2 = np.array([50, 20, 40, 20, 60, 70])

# Data for line 3
x3 = np.array([0, 1, 2, 3, 4, 5])
y3 = np.array([10, 20, 30, 40, 50, 60])

# Data for line 4
x4 = np.array([0, 1, 2, 3, 4, 5])
y4 = np.array([200, 350, 250, 550, 450, 150])

# Create a figure with four subplots arranged in a 2x2 grid
fig, axs = plt.subplots(nrows=2, ncols=2)

# Plot the data on each subplot
axs[0, 0].plot(x1, y1)
axs[0, 0].set_title('Line 1')

axs[0, 1].plot(x2, y2)
axs[0, 1].set_title('Line 2')

axs[1, 0].plot(x3, y3)
axs[1, 0].set_title('Line 3')

axs[1, 1].plot(x4, y4)
axs[1, 1].set_title('Line 4')

# Add a main title for the figure
fig.suptitle('Four Line Plots')

# Display the plots
plt.show()
```

In this code, we first define four sets of data for four different line plots. We then create a figure with four subplots arranged in a 2x2 grid using the subplot() function. We plot each set of data on a separate subplot using the plot() function and set the title for each subplot using the set_title() function. Finally, we add a main title for the figure using the suptitle() function and display the plots using the show() function.

The resulting figure will display four line plots side-by-side, each with its own title, and a main title for the entire figure.

Q4: What is a bar plot? Why is it used? Using the following data plot a bar plot and a horizontal bar plot.
import numpy as np
company = np.array(["Apple", "Microsoft", "Google", "AMD"])
profit = np.array([3000, 8000, 1000, 10000])

A bar plot is a type of plot that displays data with rectangular bars. The length of each bar is proportional to the value it represents, and the bars can be either vertical or horizontal. Bar plots are useful for comparing the relative sizes of different categories or groups of data. They can be used to display data that is discrete or categorical in nature.

Using the provided data, we can create a bar plot and a horizontal bar plot using Matplotlib. Here's the code:

```python
import numpy as np
import matplotlib.pyplot as plt

# Data for the bar plot
company = np.array(["Apple", "Microsoft", "Google", "AMD"])
profit = np.array([3000, 8000, 1000, 10000])

# Create a bar plot
plt.bar(company, profit)
plt.title('Bar Plot')
plt.xlabel('Company')
plt.ylabel('Profit')
plt.show()

# Create a horizontal bar plot
plt.barh(company, profit)
plt.title('Horizontal Bar Plot')
plt.xlabel('Profit')
plt.ylabel('Company')
plt.show()
```

In this code, we first define the data for the bar plot as two NumPy arrays: company and profit. We then use the bar() function to create a vertical bar plot and the barh() function to create a horizontal bar plot. We set the title, xlabel, and ylabel for each plot using the corresponding functions, and display the plots using the show() function.

The resulting bar plot will display the profits of each company as rectangular bars, and the horizontal bar plot will display the same information but with the bars oriented horizontally.

Q5: What is a box plot? Why is it used? Using the following data plot a box plot.
box1 = np.random.normal(100, 10, 200)
box2 = np.random.normal(90, 20, 200)

A box plot is a type of plot that displays the distribution of a dataset based on its five-number summary: the minimum value, the first quartile (Q1), the median (Q2), the third quartile (Q3), and the maximum value. It can also display outliers and skewness in the data. Box plots are useful for identifying differences and similarities between different groups of data.

Using the provided data, we can create a box plot using Matplotlib. Here's the code:

```python
import numpy as np
import matplotlib.pyplot as plt

# Data for the box plot
box1 = np.random.normal(100, 10, 200)
box2 = np.random.normal(90, 20, 200)
data = [box1, box2]

# Create a box plot
plt.boxplot(data)
plt.title('Box Plot')
plt.xticks([1, 2], ['Box 1', 'Box 2'])
plt.xlabel('Box')
plt.ylabel('Value')
plt.show()
```

In this code, we first define two sets of data as NumPy arrays: box1 and box2. We then combine them into a single list called data. We use the boxplot() function to create a box plot of the data, set the title, xlabel, and ylabel for the plot, and display it using the show() function.

The resulting box plot will display the five-number summary and any outliers for each dataset. It will also show the differences and similarities between the two datasets.
